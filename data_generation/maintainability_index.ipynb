# Initial

```python
import json

with open('initial_mi.json') as f:
    initial_mi = [item['maintainability_index'] for item in json.load(f) if item['maintainability_index'] != None]

print(initial_mi)
```

```python
import matplotlib.pyplot as plt

# Draw the distribution
plt.hist(initial_mi, edgecolor='black')
plt.title('Distribution of Initial Maintainability Index')
plt.xlabel('Maintainability Index')
plt.ylabel('Frequency')

# Save the figure to a file
plt.savefig('maintainability_index_distribution.png')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Assume maintainability_index is already defined
# maintainability_index = [...]  # This line is commented out because maintainability_index is already defined

# Draw the boxplot
plt.boxplot(initial_mi, vert=False)  # vert=False makes the boxplot horizontal
plt.title('Boxplot of Maintainability Index')
plt.ylabel('Maintainability Index')

# Save the figure to a file (optional)
plt.savefig('maintainability_index_boxplot.png')

# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assume maintainability_index is already defined
# maintainability_index = [...]  # This line is commented out because maintainability_index is already defined

# Draw the violinplot
sns.violinplot(x=initial_mi, orient='v')  # orient='v' makes the violinplot vertical
plt.title('Violinplot of Maintainability Index')
plt.xlabel('Maintainability Index')

# Save the figure to a file (optional)
plt.savefig('maintainability_index_violinplot.png')

# Show the figure
plt.show()
```

# Regenerated

```python
import json

with open('regenerated_mi.json') as f:
    mi = [item['maintainability_index'] for item in json.load(f) if item['maintainability_index'] != None]

print(mi)
```

```python
import matplotlib.pyplot as plt

# Draw the distribution
plt.hist(mi, edgecolor='black')
plt.title('Distribution of Maintainability Index of Regenerated Code')
plt.xlabel('Maintainability Index')
plt.ylabel('Frequency')

# Save the figure to a file
plt.savefig('maintainability_index_distribution.png')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assume maintainability_index is already defined
# maintainability_index = [...]  # This line is commented out because maintainability_index is already defined

# Draw the violinplot
sns.violinplot(x=initial_mi, orient='v', label='initial')  # orient='v' makes the violinplot vertical
sns.violinplot(x=mi, orient='v', label='regenerated')  # orient='v' makes the violinplot vertical
plt.title('Violinplot of Maintainability Index')
plt.xlabel('Maintainability Index')

# Save the figure to a file (optional)
plt.savefig('maintainability_index_violinplot.png')

plt.legend()

# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assume maintainability_index is already defined
# maintainability_index = [...]  # This line is commented out because maintainability_index is already defined

# Create a figure and a set of subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Draw the violinplot for the initial maintainability index
sns.violinplot(ax=axes[0], x=initial_mi, orient='v')  # orient='v' makes the violinplot vertical
axes[0].set_title('Initial Maintainability Index')
axes[0].set_xlabel('Maintainability Index')

# Draw the violinplot for the regenerated maintainability index
sns.violinplot(ax=axes[1], x=mi, orient='v')  # orient='v' makes the violinplot vertical
axes[1].set_title('Regenerated Maintainability Index')
axes[1].set_xlabel('Maintainability Index')

# Save the figure to a file (optional)
plt.savefig('maintainability_index_violinplot.png')

# Show the figure
plt.show()
```
